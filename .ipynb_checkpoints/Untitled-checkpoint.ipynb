```python
import torch
import torch.nn as nn
import torch.functional as F
import torchvision
import torchvision.transforms as transforms
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from tqdm.notebook import tqdm
import torch.cuda.amp as amp
from sklearn.manifold import TSNE

plt.rcParams['axes.labelsize'] = 18
plt.rcParams['axes.titlesize'] = 18
plt.rcParams['xtick.labelsize'] = 18
plt.rcParams['ytick.labelsize'] = 18
plt.rcParams['figure.figsize'] = (11, 7)

palette = ['#5E81AC', '#81A1C1', '#88C0D0', '#8FBCBB', '#BF616A', '#D08770', '#EBCB8B', '#A3BE8C', '#B48EAD']
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
def get_mnist_data(batch_size=8):
#     transform = transforms.Compose([
#         transforms.FiveCrop(25),
#         transforms.Lambda(
#             lambda crops: torch.stack([transforms.ToTensor()(crop) for crop in crops])
#         )
#     ])
    
    train_loader = torch.utils.data.DataLoader(
        torchvision.datasets.MNIST(
            '/data/datasets', 
            train=True, 
            download=True,                  
            transform=transforms.ToTensor()
        ),
        batch_size=batch_size,
        shuffle=True
    )

    test_loader = torch.utils.data.DataLoader(
        torchvision.datasets.MNIST(
            '/data/datasets', 
            train=False, 
            download=True,
            transform=transforms.ToTensor()
        ),
        batch_size=batch_size, 
        shuffle=True
    )
    
    return train_loader, test_loader
```

```python
trainloader, testloader = get_mnist_data(batch_size=512)
```

```python
class VAE(nn.Module):
    def __init__(self, latent_dim, image_dim):
        super().__init__()
        
        # general attributes
        assert(image_dim > latent_dim)
        self.latent_dim = latent_dim
        self.image_dim = image_dim
        self.relu = nn.LeakyReLU()
        self.drop = nn.Dropout(0.3)
        self.sigmoid = nn.Sigmoid()
        
        self.encoder = nn.Sequential(
            nn.Linear(image_dim, latent_dim ** 3),
            self.relu,
            self.drop,
            nn.Linear(latent_dim ** 3, latent_dim ** 3),
            self.relu,
            self.drop,
            nn.Linear(latent_dim ** 3, latent_dim * 2)
        )
    
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, latent_dim ** 3),
            self.relu,
#             self.drop,
            nn.Linear(latent_dim ** 3, latent_dim ** 3),
            self.relu,
            nn.Linear(latent_dim ** 3, image_dim),
        )
    
    def reparameterize(self, mu, logvar):
        if self.training:
            std = logvar.mul(0.5).exp_()
            eps = std.data.new(std.size()).normal_()
            return eps.mul(std).add_(mu)
        else:
            return mu
    
    def forward(self, x):
        mu_logvar = self.encoder(x).view(-1, 2, self.latent_dim)
#         print(mu_logvar.shape)
        mu = mu_logvar[:,0,:]
        logvar = mu_logvar[:,1,:]
        z = self.reparameterize(mu, logvar)
#         print(z.shape)
        return self.decoder(z), mu, logvar
```

```python
v = VAE(16, 28**2)
v = v.to(device)
```

```python
def plot_grid(images, nrows=1, ncols=4):
    fig = plt.figure(figsize=(ncols * 4, nrows * 4))
    for i in range(len(images)):
        ax = fig.add_subplot(nrows, ncols, i + 1)
        ax.imshow(images[i])
        ax.axis('off')
```

```python
z = torch.randn((4, 16)).to(device)
gen = v.decoder(z).view(-1, 28, 28).detach().cpu()
plot_grid(gen)
```

```python
optim = torch.optim.Adam(v.parameters(), lr=1e-3)
```

```python
def loss(xhat, x, mu, logvar, beta=1):
    reconstruction = nn.functional.binary_cross_entropy_with_logits(xhat, x.view(xhat.shape), reduction='sum')
    kld = 0.5 * torch.sum(logvar.exp() - logvar - 1 + mu.pow(2))
    return reconstruction + beta * kld
```

```python
epochs = 50

scaler = amp.GradScaler()
trainlosses = []
testlosses = []

for e in tqdm(range(epochs)):
    
    # training
    v.train()
    train_loss = 0
    for feature, _ in trainloader:
        with amp.autocast():
            feature = feature.to(device)
            output, mu, logvar = v(feature.view(-1, 28 * 28))
            l = loss(output, feature, mu, logvar)
            train_loss += l.item()
        optim.zero_grad()
        scaler.scale(l).backward()
        scaler.step(optim)
        scaler.update()

#         feature = feature.to(device)
#         output, mu, logvar = v(feature.view(-1, 28 * 28))
#         l = loss(output, feature, mu, logvar)
#         train_loss += l.item()
#         l.backward()
#         optim.step()
        
    trainlosses.append(train_loss)
    print(f'epoch {e}, training loss {train_loss / len(trainloader):.3f}')

    # testing
    with torch.no_grad():
        v.eval()
        test_loss = 0
        for feature, label in testloader:
            with amp.autocast():
                feature = feature.to(device)
                output, mu, logvar = v(feature.view(-1, 28 * 28))
                l = loss(output, feature, mu, logvar)
                test_loss += l.item()

    testlosses.append(test_loss)
    print(f'epoch {e}, testing loss {test_loss / len(testloader):.3f}')
```
